```python
from google.colab import userdata
import os

os.environ["GROQ_API_KEY"] = userdata.get("Chave_Groq")
```

```python
from groq import Groq
import os

client = Groq(api_key=os.environ.get("GROQ_API_KEY"))

response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": "daniel vorcaro banco master"}],
)

print(response.choices[0].message.content)
```

O que é o LLaMA 3.1 8B Instant
É um modelo criado pela Meta (empresa do Facebook) e rodando nos servidores da Groq. Foi lançado em julho de 2024, tem 8 bilhões de parâmetros e é otimizado para conversação com contexto de 128k tokens e capacidades de raciocínio avançadas.
