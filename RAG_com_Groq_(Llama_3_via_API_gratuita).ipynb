Carregar documento

```python
from langchain_community.document_loaders import PyPDFLoader

# Coloque exatamente o nome do arquivo
loader = PyPDFLoader("/content/Manual_Basico_do_Windows_11_v1.pdf")
docs = loader.load()
```

Dividir em chunks

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=200
)

chunks = splitter.split_documents(docs)
```

Criar embeddings

```python
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vectorstore = FAISS.from_documents(chunks, embeddings)
retriever = vectorstore.as_retriever()
```

Configurar Llama 3 via Groq

Modelo recomendado (rápido e bom):

```python
llm = ChatGroq(
    model="llama3-8b-8192",
    temperature=0
)
```

Outras opções:

"llama3-70b-8192" (melhor qualidade, mais limite)

"mixtral-8x7b-32768"

```python
llm = ChatGroq(
    model="llama3-8b-8192",
    temperature=0
)
```

Criar cadeia RAG

```python
qa = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever
)
```

Perguntar

```python
qa.run("Qual é o tema principal do documento?")
```

```python
🔥 O que está acontecendo aqui?

Pergunta →
FAISS busca os chunks relevantes →
Contexto enviado para Llama 3 →
Resposta fundamentada no seu documento

Isso é RAG.
```
